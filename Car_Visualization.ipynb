```
df = pd.read_csv('USA_cars_datasets.csv')
```

```
print(df.head(5))
print(df.columns)
```

```
#Which Brand sells the most?
sns.countplot(x = df['brand'], order = df['brand'].value_counts().iloc[:5].index, palette="husl", )
plt.xticks(rotation=40)
```

```
#Which Color is the most popular? 
sns.countplot(x = df['color'], order = df['color'].value_counts().iloc[:5].index )
plt.xticks(rotation=40)
```

```
#Which model year is most popular?
sns.countplot(x = df['year'], order = df['year'].value_counts().iloc[:5].index, palette='spring')
plt.xticks(rotation=40)
```

```
#Which brand is the most expensive?
carPrice = df.groupby('brand').price.mean().sort_values(ascending = False)
df_2 = pd.Series.to_frame(carPrice)
sns.barplot(x = df_2.index, y = df_2['price'], order = df_2.index[:5], palette = "pastel")
```

```
#Which states buy the most cars?
sns.countplot(x = df['state'], order = df['state'].value_counts().iloc[:5].index, palette='dark')
plt.xticks(rotation=40)
```

```
#Out of the top 5 buying states, which brands do they buy?
buyingStates = ['pennsylvania', 'florida', 'texas', 'california', 'michigan']
for i in buyingStates:
  carSorted = dict(df.loc[df['state'] == i]['brand'].value_counts())
  plt.bar(carSorted.keys(), carSorted.values())
  plt.xticks(rotation=90)
  plt.legend(buyingStates)
```
